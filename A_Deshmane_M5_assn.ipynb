#### 1. What proportion of diamonds are between .30 and 1.08 carats?
Count the carat values that lies between 0.30 and 1.08 and assign it to prop1

Count the total values of carat attribute and assign it to total

Divide the prop1 by total to get the proportion

Print the result

```python
prop1 = diamonds[(diamonds.carat > 0.30) & (diamonds.carat < 1.08)].carat.count()
total = diamonds['carat'].count()

prop = (prop1/total)*100

print("The percentage of diamonds between 0.30 and 1.08 is :",round(prop,2))
```

#### 2. How many of the diamonds have equal x and y dimensions?
Count the diamonds where its x and y dimensions are same

Print the result

```python
dim1 = diamonds[diamonds.x == diamonds.y]['x'].count()
print("The count of diamonds with equal x and y dimensions is :",dim1)
```

#### 3. How many of the diamonds have a carat less than the mean carat value?
Calculate the mean of carat column and assign it to mean1

Count the carat values less than mean1

Print the result

```python
mean1 = diamonds['carat'].mean()
mean2 = diamonds[(diamonds.carat < mean1)].carat.count()

print(mean2,"diamonds have a carat less than the mean carat value")
```

#### 4. How many diamonds have a Premium cut or better? Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair / Good / Very Good / Premium / Ideal
Count the diamonds with premium or ideal cut

Print the result

```python
cut1 = diamonds[(diamonds.cut == 'Premium') | (diamonds.cut == 'Ideal')].cut.count()
print(cut1,"diamonds have a premium or ideal cut")
```

#### 5. Which diamond has the highest price per carat? What is its value?
Add price per carat column to the diamonds dataset with values of price per carat

To calculate values for new column divide price by carat

To check which diamond has the highest price per carat use idxmax. It gives the index of maximum value

Check the max value of the price per carat using max function

Print the result

```python
diamonds['Price per carat']= diamonds['price']/diamonds['carat']

print("The diamond with index number",diamonds['Price per carat'].idxmax(),"has the highest price per carat")
print("The highest price per carat is",round(diamonds['Price per carat'].max(),2))
```
