Using readlines() get a list containing each line in the file

```python
with data as l:
    line = l.readlines()
line
```

Checking the type of 'line' and each index from the 'line'. The type is list of the 'line' but bytes of the line index

Convert the byte values into string using decode() and append the line index into lines list. String type is important to perform string operations on it.

```python
lines = []
for a in range(len(line)):
    lines.append(line[a].decode())
lines
```

Using regex to find Player number, Player name & Total points

```python
Player_Number = []
player_Name = []
total_Points = []
i = 4
while i < len(lines): 
    Player_Number.append(re.findall('[\d]+',lines[i])[0]) # to find player no.
    player_Name.append(re.findall('[A-Z-]+[A-Z-\s]+[A-Z-]+',lines[i])[0]) #to find player name
    total_Points.append(re.findall('[\d.]{3}',lines[i])[0]) # to find total points
    i=i+3
    
print("Player number :",Player_Number)
print("player's Name :",player_Name)
print("total Points :",total_Points)
```

###### In above code we have used 3 regular expressions are used for parsing/pattern matching within text. To extract the Player number, name and total points . we've defined a regular expressions for that:

__'[\d]+'__ --Player number regex--

The parsing behavior of this regular expression can be explained as follows:

__[\d]__ =   match any number within the range of 0,1,2,3...9.

__'+'__ =   A plus sign ('+') within a regular expression indicates that the regular expression parser should match one or more of the preceding regex characters/codes. 

---------------------------------------------------------------------

__[A-Z-]+[A-Z-\s]+[A-Z-]+__ -- Player's name regex

The parsing behavior of this regular expression can be explained as follows:


__[A-Z-]__= Match any __uppercase__ letter within the range of A, B, C, .., Z and match 1 or more occurrence of "-".

__'+'__ =   A plus sign ('+') within a regular expression indicates that the regular expression parser should match one or more of the preceding regex characters/codes. 

__[A-Z-\s]__= Match any __uppercase__ letter within the range of A, B, C, .., Z and match 1 or more occurrence of "-" and match 1 or more occurrence of white space.

__'+'__ =   A plus sign ('+') within a regular expression indicates that the regular expression parser should match one or more of the preceding regex characters/codes. 

__[A-Z-]__= Match any __uppercase__ letter within the range of A, B, C, .., Z and match 1 or more occurrence of "-".

__'+'__ =   A plus sign ('+') within a regular expression indicates that the regular expression parser should match one or more of the preceding regex characters/codes. 

The combination helps avoid extra spaces in the name.

----------------------------------------------------------------------

__[\d.]{3}__ -- Total points regex

The parsing behavior of this regular expression can be explained as follows:



__[\d.]{3}__= Match any number within the range of 0,1,2,3...9.and "." it __exactly__ can occur 3 times..

__{3}__ = Match exactly three of the preceding regex characters/codes. In this regex the preceding characters/codes are the Match any number within the range of 0,1,2,3...9 and "."





```python
State = []
Pre_rating = []
j = 5
while j < len(lines):
    State.append(re.findall('[A-Z]{2}',lines[j])[0]) # To find state
    Pre_rating.append(re.sub("R:\s+","",re.findall('R:\s+[0-9]+',lines[j])[0]))
    j=j+3 # To find Pre_rating
    
print("State :",State)
print("Pre-rating :",Pre_rating)
```

Using regex to get the Round details

```python
Round = []
k = 4
while k < len(lines):
    Round.append(re.findall('[WLDUHBX]\s+[\d]*',lines[k])[-7:]) # using regex to extract player's no. from each rounds.
    k=k+3
Round
```

###### In above code we have used 1 regular expression for parsing/pattern matching within text. To extract the rounds player number . we've defined a regular expressions for that:

__'[WLDUHBX]\s+[\d]*'__

The parsing behavior of this regular expression can be explained as follows:


__[WLDUHBX]__= Match any uppercase letter only (WLDUHBX) match exactly 1 occurrence. 

__\s__=  matches a single whitespace character -- space, newline, return, tab, form [ \n\r\t\f].

__[\d]__= Match any number within the range of 0,1,2,3...9.

__*__= match zero or more occurrence of the preceding regex characters/codes.


#### Creating list of opponents from every round 

From the given chess point table, we can find player number of the opponent for each player. We are creating a list of opponent players' numbers for every round. 

```python
Round_1 = []  # create empty lists to store each rounds data
Round_2 = []
Round_3 = []
Round_4 = []
Round_5 = []
Round_6 = []
Round_7 = []

for a in range(len(Round)):
    for b in range(len(Round[a])):    #check each rounds no. and append into the desired list.
        if b == 0:
            Round_1.append(re.sub("\s","",Round[a][b][-2:]))
        elif b == 1:
            Round_2.append(re.sub("\s","",Round[a][b][-2:]))
        elif b == 2:
            Round_3.append(re.sub("\s","",Round[a][b][-2:]))
        elif b == 3:
            Round_4.append(re.sub("\s","",Round[a][b][-2:]))
        elif b == 4:
            Round_5.append(re.sub("\s","",Round[a][b][-2:]))
        elif b == 5:
            Round_6.append(re.sub("\s","",Round[a][b][-2:]))
        else:
            Round_7.append(re.sub("\s","",Round[a][b][-2:]))

print("Round_1 :",Round_1)
print("Round_7 :",Round_7)
```

#### Finding Average Pre Tour Score:

We need to develop a function that iterates through the data and returns with the required outcome. The function returns a column with the players' average pre-tournament chess rating of opponents by indexing the cells in the columns to their respective indexed players' ratings as well as replacing those. After this replacement, all the data is computed and divided by the number of games played.

```python
def Average_Pre_Tour(score): #define a function
    player_mapping = {}       #create a dict
    average_rating = []       #create a list
    for index, row in score.iterrows():                        
        player_mapping[int(row['Player_Number'])] = int(row['Pre_rating'])# mapping players to ratings using a for loop
    for index, row in score.iterrows():
        
        if row[0] is not None:                         
            try:
                key1 = int(row.Round_1)                #Verify the opponent's player number and record their evaluations.
                row.Round_1 = player_mapping[key1]     #First round 

                key2 = int(row.Round_2)
                row.Round_2 = player_mapping[key2]     #second round

                key3 = int(row.Round_3)
                row.Round_3 = player_mapping[key3]     #third round and so no....

                key4 = int(row.Round_4)
                row.Round_4 = player_mapping[key4]    

                key5 = int(row.Round_5)
                row.Round_5 = player_mapping[key5]

                key6 = int(row.Round_6)
                row.Round_6 = player_mapping[key6]

                key7 = int(row.Round_7)                #seventh round
                row.Round_7 = player_mapping[key7]

                total = row.Round_1 + row.Round_2 + row.Round_3 + row.Round_4 + row.Round_5 + row.Round_6 + row.Round_7
                games = len([row.Round_1, row.Round_2, row.Round_3, row.Round_4, row.Round_5, row.Round_6, row.Round_7])
                average_rating.append(total/games)         # total player's rating divided by the number of games played.
            except KeyError:
                total = int(row.Round_1) + int(row.Round_2) + int(row.Round_3) + int(row.Round_4) + int(row.Round_5) + int(row.Round_6) + int(row.Round_7)
                games = len([row.Round_1, row.Round_2, row.Round_3, row.Round_4, row.Round_5, row.Round_6, row.Round_7])
                average_rating.append(total/games)
    return average_rating         #return average
```

Constructing an object that performs the "Average Pre Tour" function and returns the results in its list.

```python
#constructing an object that performs the "Average Pre Tour" function and returns the results in its list.

average_rate = Average_Pre_Tour(chess)

#A column called "Average Rating" has been added to the chess DataFrame object.
chess['Average_Rating'] = [int (n) for n in average_rate]
```
