## 1. Introduction

### What type of data is chosen for the research and why this analysis?

Through a federal-state partnership known as the Local Area Unemployment Statistics (LAUS) program, annual total estimates of the civilian labor force, employed individuals, unemployed people, and unemployment rates are generated for more than 3,000 distinct subnational regions. BLS uses models that include real-time benchmarking to the national total employment and unemployment levels from the Current Population Survey to provide state-wide estimates (CPS). The controls for substate-area estimation are these model-based estimations for the states. Under the supervision of BLS, state labor market information offices create data for substate regions utilizing standardized techniques, software systems, and certain inputs supplied by BLS.

## 2. Data Summary:
We go the data from 'Data.gov'. We found this data a good example of real world and to understand know what is happening around us

Our procedure begins by adding a few columns that we will need for our analysis. We already know that we are doing research over the period of 2017 to 2021. The "total labor", "total employed", "total unemployed", "emp rate", and "unemp rate" columns will be added.

```python
# Calculating the "total labor", "total employed", "total unemployed", "emp rate", and "unemp rate".

df['total_labor']=df.iloc[:,[3,7,11,15,19]].sum(axis=1)

df['total_employed']=df.iloc[:,[4,8,12,16,20]].sum(axis=1)

df['total_unemployed']=df.iloc[:,[5,9,13,17,21]].sum(axis=1)

df['emp_rate']=(df['total_employed']/df['total_labor']) #employment rate

df['unemp_rate']=(df['total_unemployed']/df['total_labor']) #unemployment rate

df
```

## 4. Inference:

##### 1. Find which County has highest employment rate per year.

```python
#calculate the highest employment rate for each year, use the max function and
#Print the result

HighERperCounty = df[['Employment_rate_2017','Employment_rate_2018','Employment_rate_2019','Employment_rate_2020','Employment_rate_2021']].max()
HighERperCounty
```

```python
#The index for highest employment rate for each year can be returned using idxmax function

#Get the County name for those index


y = df[['Employment_rate_2017','Employment_rate_2018','Employment_rate_2019','Employment_rate_2020','Employment_rate_2021']].idxmax()

countyName = df['Area_name'][y]
countyName
```

Per the result, "Loving county" from Texas holds the highest employment rate which is 99.2 for the year 2019



###### 2. Find the percentage of 2021 Employment rate which are between 80 and 90.

```python
a = df.Employment_rate_2019[(df['Employment_rate_2021']>80) & (df['Employment_rate_2021']<90)].count()

# Now, we will calculate the overall number of all values present in 'Employment_rate_2019' column to successfully create the proportion:

b = df.Employment_rate_2019.shape[0]

# once we are done with finding a and b we will divide them and multiply by 100 to get our percentage:

r = (a/b*100)

print('The percentage of Employment_rate_2021 which are between 80 and 90 is:\n',r)
```

###### 3. Show year-wise labor force trend. Make statement about the trend you get



```python
#Calculate the labor force for year using sum function

#Divide the result by 1 million to make it easy to read figure and print the result

Labor_force = df[['Civilian_labor_force_2017','Civilian_labor_force_2018','Civilian_labor_force_2019','Civilian_labor_force_2020','Civilian_labor_force_2021']].sum(axis = 0,numeric_only = True)
print("The Labor force(in million) across USA for each year is :")
print(Labor_force/1000000)
```

```python
#Define a new figure and specify that only a single subplot will be placed within the figure
#Plot the figure with green dashed line and marker o
#Provide the labels and plot the graph


fig = plt.figure()
linplt = fig.add_subplot(1, 1, 1)
linplt.plot(Labor_force, color='g', linestyle='dashed',marker ='o' )
labels = linplt.set_xticklabels(['2017','2018','2019','2020','2021'],rotation = 45, fontsize='small')

```

With 164.9 million people working, the labor force peaked in 2019 and declined to 161.2 million people in 2020.

###### now we will Plot the line graph using seaborn and set the labels for same graph.

###### 4. From total labor force, what is the proportion of the 5 states with lowest/highest employment rates?



```python
#Check the total labor force state-wise
#using groupby function.

Labor_force=df.groupby(df['State'])[['Civilian_labor_force_2017','Civilian_labor_force_2018','Civilian_labor_force_2019','Civilian_labor_force_2020','Civilian_labor_force_2021']].sum().sum(axis = 1)
Labor_force
```

```python
Labor_force = Labor_force.sort_values()
Labor_force
```

Data were sorted to get the greatest and lowest values, and they were then saved to the same variable.

California has the greatest work force, while Wyoming has the lowest.

```python
#To get the 5 states with lowest labor force give index 0:5 and print the result
#Calculate the percentage of these by dividing it from the total labor force

Labor_force[0:5]
print(Labor_force[0:5])

LowER = (Labor_force[0:5].sum())*100/Labor_force.sum()
print(round((LowER),2))
```

"Wyoming", "Vermont", "Alaska", "Washington DC" and "Nevada" have the 5 lowest labor force

These 5 states holds only '1.11' of the total USA labor force.



```python
#To get the 5 states with higest labor force 
#Calculate the percentage of these by dividing it from the total labor force

Labor_force[-5:]
print(Labor_force[-5:])

HighER = (Labor_force[-5:].sum())*100/Labor_force.sum()
print(round((HighER),2))
```

California, Texas, Florida, New York and Pennsylvania holds for 5 highest labor force

These 5 states holds 36.44 of the total USA labor force

###### 7. Compare among NY, NJ & CT - Employment rate and Unemployment.

```python
#using pivot table to find employment rate
tristate_emp=df.query('State == ["NY", "NJ", "CT"]').pivot_table(
        columns=['State'],
        values='emp_rate',
        aggfunc='mean',
    fill_value=0)
tristate_emp
```

We have compared employment rate for NY, NJ & CT

NY & CT has almost same employment rate of 95%. NJ employment rate is slightly less, 94%.


###### Now we will use Matplotlib Plot for NY, NJ & CT employment rate comparison:

```python
#plt.rcParams["figure.figsize"] = [10, 6]
tristate_emp.plot.bar(grid=True)
```

Graph shows that NJ had the low employment rate as compare to CT and ny.

```python
#using pivot table to find unemployment rate

#Using query() function to filter states

tstate_unemp=df.query('State == ["NY", "NJ", "CT"]').pivot_table(
        columns=['State'],
        values='unemp_rate',
        #index=['County'],
        aggfunc='mean',
    fill_value=0)
tstate_unemp
```
