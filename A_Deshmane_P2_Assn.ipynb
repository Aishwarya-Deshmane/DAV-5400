### 2. Using your web scraping skills to extract data from a Katz School web page. Specifically, you will be extracting specific content from the Katz School’s “Staff” information page.



We're going to utilize a beautiful soup package to answer this topic.

A Python package called Beautiful Soup was created for jobs with short turnaround times, including screen scraping. It is effective due to three aspects:

As a toolbox for breaking down a text and extracting what you need, Beautiful Soup offers a few straightforward techniques and Pythonic idioms for navigating, finding, and changing a parse tree. Writing an application just requires a small amount of code.

Outgoing documents are automatically converted to UTF-8 by Beautiful Soup, and incoming documents are converted to Unicode. Unless the document doesn't mention an encoding and Beautiful Soup is unable to identify one, you don't need to consider encodings. The original encoding must then be specified.

With Beautiful Soup, you may experiment with various parsing techniques or trade off speed for flexibility on top of well-known Python parsers like lxml and html5lib.

```python
# load Request,re.
# load bs4 module from BeautifulSoup

import requests
import re 
from bs4 import BeautifulSoup
import pandas as pd # load pandas
import sys          # load sys


#requests module allows you to send HTTP requests using Python. 
#The HTTP request returns a Response Object with all the response data (content, encoding, status, etc).

page = requests.get("https://www.yu.edu/katz/staff")

page.content  #show the content of the page
soup = BeautifulSoup(page.content, 'html.parser') #We receive a BeautifulSoup object after passing the "Katz staff" HTML through Beautiful Soup, 
#which represents the HTML as a hierarchical data structure:

table = soup.find('div', attrs={'class':'text-only'}) #find all the data in class"text-only"
staff  = table.find('p')                         #then collect all data with the <p> tag and save it in the table body variable.
scrap = []

output_dict = []   #create an empty list.
office_name = "Office of the Dean"                
for i in soup.find_all('h3'):               #find all headers "h3"
    for sib in i.next_siblings:             #using for loop fetch all staff data from <p> tag
        if sib.name == 'p':                 #The nextSibling property is used to return the next node
                                            #of the specified node as Node object or null if the specified node is the last one
            output = sib.text + ' | ' + office_name 
            # output = output.replace('\n\n', ',')
            output = output.replace('\n', ',')      #replace all new line data with"," and junk value with" ".
            output = output.replace('\xa0', '')
            output= output.split(',,')             # split the data 
            
            for sp in output:
                pat=re.compile('([a-zA-Z/.0-9-]+@yu.edu)') #Create this regex to extract only the staff members' email addresses.
                email=re.findall(pat,sp)
                if not email:
                    email = 'N/A'                           #add condition if email not found type "n/a" in the text field.

                pat=re.compile('([0-9]{3}-[0-9]{3}-[0-9]{4})')#Create this regex to extract only the staff members' phone no.
                phone=re.findall(pat,sp)

                if not phone:                               # add condition if phone no. not found type "n/a" 
                    phone = 'N/A'
                
                temp = sp.split(',')                        #split the data and store into temp
                                                            #assign temp[0] as name and temp[1] as title
                name = temp[0]

                title = temp[1]

                output_dict.append({'name':name, 'office' : office_name,'email': email,'phone': phone,'title': title})
        if sib.name == 'h3':
            office_name = sib.text         #append the data into output_dict as dictionary 

#output_dict = [i for n, i in enumerate(output_dict) if i not in output_dict[n + 1:]]

staff_info = pd.DataFrame(output_dict[0:16])           #Create a Pandas dataframe named staff_info for output_dict  


staff_info  #show dataframe
```

###### In above code we have used 2 regular expressions are used for parsing/pattern matching within text. To extract the phone no. and email address . we've defined a regular expressions for that:

__'([a-zA-Z/.0-9-]+@yu.edu)'__ --Email Address regex--

The parsing behavior of this regular expression can be explained as follows:

__[A-Za-z/.0-9-]__ =   Match any uppercase/lowercase letter within the range of A, B, C, .., Z and a, b, c....z.

and match 1 or more occurrence of "." and match any number within the range of 0,1,2,3...9.and match 1 or more occurrence of "-".

__'+'__ =   A plus sign ('+') within a regular expression indicates that the regular expression parser should match one or more of the preceding regex characters/codes. 


__@yu.edu__ = Match __exactly__ can occurs at ends. 


----------------------------------------------------------------------


__"([0-9]{3}-[0-9]{3}-[0-9]{4})"__ --PHONR NO. REGEX

The parsing behavior of this regular expression can be explained as follows:

__[0-9]{3}__ = Match any number within the range of 0,1,2,3...9.and __exactly__ can occur 3 times..

__{3}__ = Match exactly three of the preceding regex characters/codes. In this the preceding characters/codes are the Match any number within the range of 0,1,2,3...9

__-__ = Then Match __exactly__ can occurs after 3 numbers.

__[0-9]{3}__ = Match any number within the range of 0,1,2,3...9.and __exactly__ can occur 3 times..
