### Data Preparation:

Added date column by combining year, month and day columns using parse dates and keep_date_col True to keep other columns in the dataset

Printed the subset of the data

```python
flights_data = pd.read_csv(filename,parse_dates={'Date':['Year','Month','DayofMonth']},keep_date_col=True)

flights_data.head()
```

Created a dataframe called DataCount to capture count of Early depature, On time departure and Delay departure from the whole data and assigned index for each value
 
Plotted the Pie chart to get the visual picture of DataCount

Per chart, 51% flights are delayed, 41% flights departed early and only 8% flights departed on time

```python
DataCount= pd.DataFrame({'CountOfDelay':[flights_data[flights_data.DepDelay<0]['DepDelay'].count(),
                                    flights_data[flights_data.DepDelay==0]['DepDelay'].count(),
                                    flights_data[flights_data.DepDelay>0]['DepDelay'].count()]},
                       index=['Early departure','On time departure','Delay departure '])

fplot1 = DataCount.plot.pie(y ='CountOfDelay',autopct = '%2.0f%%',figsize=(12,7));
```

Created a subset of flights_data using iloc by providing the only required columns

Captured 10027 records of the data with 0 cancelled flights and delayed depature greater than 0

Reset the index column and print the data

```python
flightsdata = flights_data.iloc[:,[0,4,7,14,15,16,8,9,12,13,11,5,6]][(flights_data.Cancelled == 0)&(flights_data.DepDelay>0)]

flightsdata.reset_index(drop=True)
```

Created DataMedian dataframe to find Median, Mean and Count of departure delays for weekdays

Plotted Bar chart to analyze Mean and Median of the Departure delays

Mean of Week 1, 2, 4 and 5 are greater than the mean of DepDelay column from flightsdata

Mean and median values follow the similar trend

Week 3 has lowest mean and median whereas Week 4 has the highest values

```python
weekdays = sorted(flightsdata['DayOfWeek'].unique())

DataMedian= pd.DataFrame({'Week':weekdays,
                         'Median':[flightsdata[flightsdata.DayOfWeek==p]['DepDelay'].median() for p in weekdays],
                         'Mean':[flightsdata[flightsdata.DayOfWeek==p]['DepDelay'].mean() for p in weekdays],
                         'Count':[flightsdata[flightsdata.DayOfWeek==p]['DepDelay'].count() for p in weekdays]})

fbarplot = DataMedian.plot.bar(x ='Week',y =['Mean','Median'],figsize=(12,7));
```

Created DataAirlines dataframe to find Median, Mean and Count of departure delays for carriers

Plotted Bar chart to analyze Mean and Median of the Departure delays

Plot shows EV carrier has the highest mean and median while CO has lowest values

```python
airlines = sorted(flightsdata['UniqueCarrier'].unique())

DataAirlines= pd.DataFrame({'Airlines': airlines,
                         'Median':[flightsdata[flightsdata.UniqueCarrier==p]['DepDelay'].median() for p in airlines],
                         'Mean':[flightsdata[flightsdata.UniqueCarrier==p]['DepDelay'].mean() for p in airlines],
                         'Count':[flightsdata[flightsdata.UniqueCarrier==p]['DepDelay'].count() for p in airlines]})

fbarplot1 = DataAirlines.plot.bar(x ='Airlines',y =['Mean','Median'],figsize=(12,7));
```

Created a dataframe called DataAirports to capture count of Early depature, On time departure and Delay departure from IAH and HOU airports separately. Assigned index for each value

Plotted the Pie chart to get the visual picture of DataCount for both the Origin airports

Per chart, 46% flights are delayed from IAH but 66% from HOU

45% flights departed early from IAH and 26% from HOU

8% flights departed on time from both the airports

```python
DataCount = pd.DataFrame({'IAH':[flights_data[(flights_data.DepDelay<0) & (flights_data.Origin=='IAH')]['DepDelay'].count(),
                                    flights_data[(flights_data.DepDelay==0) & (flights_data.Origin=='IAH')]['DepDelay'].count(),
                                    flights_data[(flights_data.DepDelay>0) & (flights_data.Origin=='IAH')]['DepDelay'].count()],
                         'HOU':[flights_data[(flights_data.DepDelay<0) & (flights_data.Origin=='HOU')]['DepDelay'].count(),
                                    flights_data[(flights_data.DepDelay==0)& (flights_data.Origin=='HOU')]['DepDelay'].count(),
                                    flights_data[(flights_data.DepDelay>0) & (flights_data.Origin=='HOU')]['DepDelay'].count()]},
                       index=['Early departure','On time departure','Delay departure '])

fplot2 = DataCount.plot.pie(subplots = True,autopct = '%2.0f%%',figsize=(12,7));
```

Created a dataframe called DataOrigin to capture mean and median of departure delay from IAH and HOU airports separately

HOU and IAH both have almost same mean and median departure delays with IAH slightly lagging than HOU
